```python
df11=df1[(df1.TotalCharges<upper_limit3) & (df1.TotalCharges>lower_limit3) ]
df11
```

```python
df12=df11[(df11.MonthlyCharges<upper_limit3) & (df11.MonthlyCharges>lower_limit3) ] 
df12
```

```python
df13=df12[(df12.tenure<upper_limit4) & (df12.tenure>lower_limit4)]
df13
```

```python
df13.replace('No phone service','No', inplace = True )
df13.replace('No internet service','No', inplace = True )
df13
```

```python
def print_unique_col_values(df13):
       for column in df13:
            if df13[column].dtypes=='object':
                print(f'{column}: {df13[column].unique()}')
```

```python
print_unique_col_values(df13)
```

```python
yes_no_columns = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup',
                  'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
for col in yes_no_columns:
        df13=df13[col].replace({'Yes': 1,'No': 0},inplace=True)
```

```python
print_unique_col_values(df13)
```

```python
df13.replace('Female',1, inplace = True )
df13.replace('Male',0, inplace = True )
print_unique_col_values(df13)
```

```python
df13.drop('customerID',axis=1,inplace=True)
```

```python
df13.isna().sum()
```

```python
#onehotencoding
df21 = pd.get_dummies(data=df13, columns=['InternetService','Contract','PaymentMethod'])
df21
```

```python
df21.dtypes
```

```python
x=df21.drop('Churn',axis=1)
y=df21['Churn']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=5)
```

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(26, input_shape=(27,), activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])
model.compile(optimizer='adam',loss ='binary_crossentropy', metrics=['accuracy'])
model.fit(x_train,y_train,epochs=100)
```

```python

model.evaluate(x_test, y_test)
```
